```python
def getDFDateSales(file, column, rowToFind):
    le = preprocessing.LabelEncoder()
    rows = le.fit(file[column]).classes_
    mydf = pd.DataFrame(columns = ['date', 'sales'])
    df = file[file.store_nbr == 1]
    for row in rows:
        if row == rowToFind:
            df = df[df[column] == row]
            print(df)
            le.fit(file.date)
            x = 0
            print(len(le.classes_))
            for date in le.classes_:
                if x < len(le.classes_) - 1:
                    mydf.loc[len(mydf.index)] = [date, df['sales'].loc[df.index[x]]]
                x += 1
    return mydf

mydf = getDFDateSales(train, 'family', 'BREAD/BAKERY')
```

```python
def deleteUnfoundDate(df1, df2):
    rows1 = df1['date']
    x = 0
    for row1 in rows1:
        rows2 = df2['date']
        for row2 in rows2:
            cond = row1 == row2
            if cond:
                df2 = df2.drop(df2[df2['date'] == row2].index)
                break
        if not cond:
            df1 = df1.drop(df1[df1['date'] == row1].index)
            x += 1
    print('there is ' + str(x) + ' unequals date')
    return df1    
        

mydf = deleteUnfoundDate(mydf, oil_df)
```

```python
print(len(mydf['date']))
```

```python
def addColumnFromDFToDFWithEqualDate(_to, _from, column):
    rows = _to['date']
    for row in rows:
        from_index = _from[_from['date'] == row].index[0]
        to_index = _to[_to['date'] == row].index[0]
        _to.at[to_index, column] = _from.at[from_index, column]
    return _to

mydf = addColumnFromDFToDFWithEqualDate(mydf, oil_df, 'dcoilwtico')
```

```python
def isValInArray(val, array):
    lenAr = len(array)
    index = 0
    while index < lenAr:
        if val == array[index]:
            return True
        index += 1
    return False

def transformStringToInt(df, exceptions):
    columns = holidays_df.columns
    for col in columns:
        if not isValInArray(col, exceptions):
            le.fit(df[col])
            df[col] = le.transform(df[col])
    return df

holidays_df = transformStringToInt(holidays_df, ['date'])
holidays_df.head()
```
